Voici la simulation de la question 6


```r
theta = 3

set.seed(43) 
Nattempts = 1000
nsample = 100
pois_sample <- c()

for (i in 1:Nattempts)  # can be written without the for loop (nicer) ! But I don't know how
{
    pois_sample <- append(pois_sample, sqrt(nsample)*(mean(rpois(nsample, theta))-theta)/sqrt(mean(rpois(nsample, theta))))
}  

qqnorm(pois_sample, pch = 1, frame =FALSE)
qqline(pois_sample, col = "steelblue", lwd = 2)
```
